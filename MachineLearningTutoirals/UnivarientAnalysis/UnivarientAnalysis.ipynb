```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

  1.CATEGORICAL DATA

a.Countplot

```python
sns.countplot(x=df['Survived'])
```

2.PIE CHART

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

2.NUMERICAL DATA

1.histogram


```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
#bins can also be used
```

b.displot

```python
sns.displot(data=df, x='Age', kde=True)
```

c.BoxPlot

```python
#we can get idea of how much outliers are present in code by plotting box plot
sns.boxplot(data=df, x='Age')
```
